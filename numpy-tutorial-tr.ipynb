## Diziden eleman seçmek, parçalamak: <a id="eleman_secme"></a>

<img src="indexing.jpg">

* `a[n]` : n. eleman.
* `a[n<x]` : n < x gibi mantıksal operatörler ile eleman seçebiliriz.
* `a[x:y]` : x ve y arasındaki elemanlar. 
* `a[::-1]` : geriye doğru tüm elemanlar.

**Tek boyutlu diziler:** <a id="tekboyutlu"></a>

````python
a = np.arange(10)
print("ilk eleman: {}, son eleman: {}".format(a[0],a[-1]))
````

````python
print("a dizisinin çift elemanları: {}".format(a[a % 2 == 0 ]))
````

````python
print("3-7 arası: {}".format(a[3:7]))
````

````python
print("Sondan başa: {}".format(a[::-1]))
````

**Çok boyutlu diziler:** <a id="cokboyutlu"></a>

* `a[n,:]` : n. satırın tamamı.
* `a[:,n]` : n. sütunun tamamı.
* `a[-1]` : Son satır.
* `a[:,-1]` : Son sütun. 

````python
a = np.arange(12).reshape(3,4)
a
````

````python
print("1. Satır: {}".format(a[1,:]))
````

````python
print("2. Sütun: {}".format(a[:,2]))
````

````python
print("Son satır: {}".format(a[-1]))
````

````python
print("Son sütun: {}".format(a[:,-1]))
````

**Not:**
``` 
":" yerine "..." kullanabiliriz yani,
arr[:,:,:,3] --> arr[...,3]
arr[3,:,:,:] --> arr[3,...] kullanabiliriz. 
```

## Şekil değiştirmek: <a id="sekil_degistirme"></a>

* `a.ndim` : dizinin boyut sayısı
* `a.size` : dizinin eleman sayısı

- `a[n]` : Dizinin n. elemanı 
- `a[n].shape` : n. elemanın biçimi
- `a.shape` : Dizinin biçimi
- `a.shape[n]` : Dizinin biçiminin n. elemanı 

- `a.reshape(r,c)` : Dizinin biçimlendirilmiş halini döndürür. **reshape yaparken rxc'nin dizinin eleman sayısına eşit olması gerekir.**
- `a.resize(r,c)` : Diziyi biçimlendirir.

- `a.ravel()` : Diziyi düz bir şekilde döndürür.
- `a.flatten()` : Diziyi düz bir kopyasını döndürür.
- `np.flip(a)` : A dizisini ters döndürür.

````python
a = np.floor(10 * rg.random((3,4)))
a
````

````python
print("Dizinin ilk elemanı: {}".format(a[0]))
print("İlk elemanın biçimi: {}".format(a[0].shape))
print("Dizinin biçimi: {} (satır,sütun)".format(a.shape))
print("Dizinin biçiminin ilk elemanı: {}".format(a.shape[0]))
print("Dizinin boyut sayısı: {}".format(a.ndim))
print("Dizinin eleman sayısı: {}".format(a.size))
````

````python
a = np.arange(10).reshape(2,5)
a
````

````python
a1 = a.flatten()
a1[0] = 100 #orijinal diziyi değiştirmez.
a
````

````python
a2 = a.ravel()
a2[0] = 100 #orijinal dizi değişir
a
````

````python
np.flip(a)
````

## np.ix_() fonksiyonu: <a id="npix"></a>

* 1 boyutlu N tane diziyi, N adet N boyutlu diziye çevirir. Her dizi için N boyuttan sadece biri birim değildir.
* A + Bx gibi bir eşitliği hesaplamak için kullanılabilir.

````python
a = np.arange(10).reshape(2, 5)
a
````

````python
ixgrid = np.ix_([0, 1], [2, 4])
ixgrid
````

````python
ixgrid[0].shape, ixgrid[1].shape
````

````python
a[ixgrid]
````

*A + Bx (result) kod örneği:*

````python
a=np.array([2,3,4,5])
b=np.array([8,5,4])
c=np.array([5,4,6,8,3])
ax,bx,cx=np.ix_(a,b,c)
````

````python
print (ax , "--- ax'in biçimi: {}".format(ax.shape))
````

````python
print (bx , " --- bx'in biçimi: {}".format(bx.shape))
````

````python
print (cx , "--- cx'in biçimi: {}".format(cx.shape))
````

````python
result = ax + bx * cx
result
````

````python
result [2,0,4]
````

````python
a[2] + b[0] * c[4]
````

## Lineer Cebir ve Numpy <a id="lineer_algebra"></a>

* `a.transpose()` : matrisin transpozunu döndürür.
* `np.linalg.inv(a)` : matrisin tersini(inverse) alır.
* `np.eye(a)` : axa'lık birim matrix --> "I"
* `np.trace(a)` : a matrisinin özdeğerini hesaplar.
* `np.linalg.eig(a)` : a(kare matris) matrisinin özdeğerini ve özvektörünü döndürür.

````python
a=np.array([[1.0,2.0], [3.0,4.0]])
print(a)
````

````python
a.transpose()
````

````python
np.linalg.inv(a)
````

````python
u=np.eye(2)
u
````

````python
j=np.array([[0.0,-1.0], [1.0,0.0]])
````

````python
j@j #matris çarpımı
````

np.trace(u) #özdeğer

````python
y=np.array([[5.], [7.]])
np.linalg.solve(a, y)
````

````python
np.linalg.eig(j) 
````

* `np.unique(a)` : a dizisindeki elemanları tekrarsız döndürür.

````python
a=np.array([11,11,12,13,14,15,16,17,12,13,11,14,18,19,20])
unique_values=np.unique(a)
print(unique_values)
````

````python
unique_values, indices_list=np.unique(a, return_index=True) #tekrarlayan elemanın bulunduğu ilk indis.
print(indices_list)
````

````python
unique_values, occurrence_count=np.unique(a, return_counts=True) #elemanların tekrar sayısı.
print(occurrence_count)
````
